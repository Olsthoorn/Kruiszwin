# shapefile.Reader reads the three files making up the `shapefile`

A shapefile consists of three files with extensions `.shp`, `.shx`, `.dbf`. The file with extension `.shx` is a pointer file that is used for access efficiency only; it's not required.

```python
# Switch to the directory and open the shapefile
cwd = os.path.expanduser('~/GRWMODELS/python/tools/shape/shapefiles')
os.chdir(cwd)

# Open a shapefile
sf = shapefile.Reader(os.path.join(cwd, "Breuken"))
```

```python
# Than plot the lines in it
ax = newfig("Breuken in Limburg", "xRD [m]", "yRD [m]", size_inches=(6, 10))
for shp in sf.shapes():
    xy = np.array(shp.points)
    ax.plot(*xy.T)
ax.set_aspect(1)
```

This plotting is not the whole story as the lines thus obtained may, in fact, be multilines. This is defined by all multilines beeing a single list of coordinate pairs, while the associated `parts` determine the beginning of each subline within the total list of coordinates. This implies, a multiline can be read as a single line, but without the parts it would miss the breaking up into individual lines.

Print the open shapefile, this yields number of shapes and records and the number of fields.

The number of fields concludes the first one with the "Deletionflag", which should be ignored. It seems to be an option within use of `.dbf` files. Therefore, there are actually only n-1 real fields determined by the user.

Therefore, the points should be treated as follows.

Here is a way to deal with multlines by making use of the `.parts` and the `.points` attributes of a shape.

```python
print("The `.parts` are s shapefile._Array not a list, so use hstack to add the end point (+1)")

prt = np.hstack((sf.shape(1).parts, len(sf.shape(1).points)))

print(sf.shape(1).parts, type(sf.shape(1).parts))
print(prt, type(prt))
```

Combining `.points` and `.parts` to regenerate a multiline

```python
line = []
for i,j in zip(prt[:-1], prt[1:]):
    line.append(sf.shape(1).points[i:j])

print('Multiline of shape(1):')
pprint(line)
print('\nFirst line of shape(1):')
pprint(line[0])
print('\nSecond line of shape(1):')
pprint(line[1])
```

When we store the lines in the shape in our own object (`Shape_df` object in module `tools.shape.shapetools`), we store the lines in `path.Path` objects and the records in a `pd.DataFrame`. It's easy to generate a path.Path object from the points.

```python
path.Path(sf.shape(1).points)
```

However, now we lost our multiline information. Multiline information can easily be included in the path by using codes, which are None above. Codes is an array of integers telling when a line has to be drawn from the current pen location to the actual coordinate or whether the pen has te lifted, i.e. moved to this location. So we have to generate an array of integers with same length as the number of points in which each point to be connected from the previous get value LINETO (=2) and when it's not te be connected the value MOVETO (=1). There are other options as well. A way to do this using the points and parts is as follows.

```python
sf.shape(1).parts
```

```python
MOVETO = 1
LINETO = 2
paths = []
for shp in sf.shapes():
    pnts = shp.points
    prts = shp.parts
    codes = LINETO * np.ones(len(pnts), dtype=int)
    codes[prts] = MOVETO
    paths.append(path.Path(pnts, codes=codes))
paths[:2]
```

## Fields

Let's first look at the first three fields:

```python
sf.fields[:3] # The first field is not used, so ignore it.
```

The first field is a tuple, with "DeletionFlag" as first argument. This can be ignored. This makes the number of fields one longer than the len of each of the records.
Each field has a name, type, width and an accuracy. The types are 'C' for character, 'N' for number.

Hence we will always ignore the first field of a given shapefile. This "Deletionflag" is nog in the tutorial of `pyshp`; it seems to have to do with an option for uses of the `.dbf` file.

When generating fields to be saved to disk, we don't generate this `Deletionflag` field. at all.

Next we extract the different items from the fields while ignoring the first field

```python
# Possible types are 'N' or 'F' for a number, 'L' for logical, 'D' for date and 'C' for object, which will be a string
columns = [fld[0] for fld in sf.fields][1:]
types   = [fld[1] for fld in sf.fields][1:]
length  = [fld[2] for fld in sf.fields][1:]
decimal = [fld[3] for fld in sf.fields][1:]

#print(columns)
print(types)
#print(length)
#print(decimal)
```

The types can be `L`, `N` or `F`, `D`, `C` for logical, a number (int or float), a date as `yyyymmdd` or an object (always a string). To discern a floating point from an integer number, use nonzero or zero for the decimals field in each field. To accomodate number, the length field must suffice to store all the digits in front of the decimal point of the floating point number (or all the digits of an integer). Therefore, 1e20, requires 21 for the length field. The length and decimals do not matter for logicals and strings.

When generating a pd.DataFrame, one can only use a single dtype in its call. Therefore, generally, the only option is that dtypes of DataFrame columns are infered by `pandas`. However, we need to specify fields when generating them for shapefile.Writer. A way to do this is the following.

```python
types=[]
for fld in sf.fields[1:]:
    if fld[1] in 'NF':
        if fld[3] == 0:
            types.append(np.int64)
        else:
            types.append(np.float64)
    elif fld[1] == 'D':
        types.append(date)
    elif fld[1] == 'L':
        types.append(bool)
    else:
        types.append(str)
for i, (nm, tp) in enumerate(zip(sf.fields[1:], types)):
    print('{:30s} {:30s}'.format(str(nm), str(tp)))
    if i == 3:
        break   
```

The other way around is more important, i.e. generating shapefile fields given the actual dtypes of the columns of a pd.DataFrame.

```python
def dbf_type(self, tp, strlen=30, digits=10, decimals=10):
        """Return dbf field header with type and leng and decimals.
        
        Parameters
        ----------
        tp: a np.type
            type like int, float, str, np.dtype('O'), np.dtype('M<8'), np.datetime64
        strlen: int
            length of strings to be used when generating a shapfile string field.
        digits: int
            length of digits before decimal point to be used for both float and ints
        decimals: int
            the number of decimals to be used for floating poitns (0 for ints)
        """
        if tp == bool:
            return ('L', 1, 0)
        elif tp == int:
            return ('N', digits, 0)
        elif tp == float:
            return ('N', digits, decimals )
        elif tp == np.dtype('O') or tp == str:
            return ('C', strlen, 0)
        elif tp == np.datetime64:
            return ('D', 8, 0)
        else:
            raise ValueError('Unknown dtype: {}. May be cast it to a known data type'.format(tp))
```

```python
columns = [fld[0] for fld in sf.fields[1:]]
data = pd.DataFrame([list(rec) for rec in sf.records()], columns=columns)
data
```

Now see if we gen generate a `Shape_df` object using the paths and a pd.DataFrame assembled above.

In case we need a bit more space, we may well specify our own `xlim` and `ylim` instead of using the internal `.bbox` automatically. Or it can be specified using the internal faults.bbox and passing a `rounder`, which will determine `xlim` and `ylim` using the internal `bbox` but rounded to multiples of `rounder`.

```python
faults = sht.Shape_df(paths=paths, data=data) # generate a Shape_df object from paths and a DataFrame
ax = newfig("Faults", "xRD", "yRD")           # New figure
faults.plot(ax=ax, rounder=10000)             # Plot, using internal bbox rounded to rounder
```
