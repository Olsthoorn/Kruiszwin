```python
def newfig(title=None, xlabel=None, ylabel=None, xlim=None, ylim=None,
           xscale=None, yscale=None, size_inches=(14, 6)):
    """Generate a standard new figure."""
    fig, ax = plt.subplots()
    fig.set_size_inches(size_inches)
    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    ax.grid()
    if xlim   is not None: ax.set_xlim(xlim)
    if ylim   is not None: ax.set_ylim(ylim)
    if xscale is not None: ax.set_xscale(xscale)
    if yscale is not None: ax.set_yscale(yscale)
    return ax
```

```python
pnames = ['B14B0131001_1.csv', 'B14B0131002_1.csv', 'B14B0155001_1.csv']
skiprows = [16, 16, 15]
seps = ['/', '-', '-']

pb=dict()

for bname, skrws, sep in zip(pnames, skiprows, seps):
    print(bname)
    pb[bname] = pd.read_csv(os.path.join(csvdir, bname),
        skipinitialspace=True,
        skiprows=skrws,
        usecols=[2, 5],
        verbose=False,
        parse_dates=True,
        dayfirst=True,
        decimal= '.',
        error_bad_lines=True,
        warn_bad_lines=True,
        delim_whitespace=False,                            
    )
    pb[bname].index=[np.datetime64('-'.join(p.split(sep)[::-1])) for p in pb[bname]['Peildatum']]
    print(pb[bname].columns)
    print(pb[bname].dtypes)
    #break
    #pb[bname].index = pb[bname]['Peildatum']
```

```python
ax = newfig("Peilbuizen kruiszwin met langjarige reeks", "tijd", "cm +NAP")
for k in pb:
    pf = pb[k]
    ax.plot(pf.index, pf['Stand (cm t.o.v. NAP)'], label=bname)
ax.legend()
```
