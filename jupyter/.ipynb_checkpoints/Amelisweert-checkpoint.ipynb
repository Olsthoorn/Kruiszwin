```python
def newfig(title=None, xlabel=None, ylabel=None,
           xscale=None, yscale=None, xlim=None, ylim=None,
           size_inches=(14, 7)):
    """Setup a new figure."""
    fig, ax = plt.subplots()
    fig.set_size_inches(size_inches)
    if title: ax.set_title(title)
    if xlabel: ax.set_xlabel(xlabel)
    if ylabel: ax.set_ylabel(ylabel)
    if xscale: ax.set_xscale(xscale)
    if yscale: ax.set_yscale(yscale)
    if xlim: ax.set_xlim(xlim)
    if ylim: ax.set_ylim(ylim)
    ax.grid()
    return ax
```

# Hantush well function and checking it

```python
def Wh(u, rho, n=5000):
    """Return drawdown according to Hantush for a time sequence."""
    if not np.isscalar(rho):
        raise ValueError('Rho must be scalar.')
    if np.isscalar(u):
        u = np.array([u])
    eps, inf = 1e-100, 20.0
    # generate sufficient u values for integration and mix with the given ones
    # The eps ensures that we start integration immediately after each u
    y = np.hstack((u + eps, np.logspace(np.log10(np.min(u)), inf, n)))
    y.sort()
    a = np.exp(-y - (rho / 2) ** 2 / y) / y
    dy = np.diff(y)
    dy2 = np.zeros_like(y)
    dy2[:-1] += 0.5 * dy
    dy2[1: ] += 0.5 * dy
    # Rapid filling of the answer array
    w = np.zeros_like(u)
    for i, uu in enumerate(u):
        w[i] = np.sum(a[ y >= uu] * dy2[y >= uu])
    return w
```

```python
from scipy.special import exp1

fig, ax = plt.subplots()
fig.set_size_inches(14, 8)
ax.grid()
ax.set_title("Hantush type curves")
ax.set_xlabel('1/u')
ax.set_xscale('log')
ax.set_yscale('log')
u = np.logspace(-6, 0, 71)

ax.plot(1/u, exp1(u), '.', label="Theis")
for rho in [0.01, 0.03, 0.1, 0.3, 1]:
    ax.plot(1/u, Wh(u, rho), label='r/L={:.3g}'.format(rho))
ax.legend()
```

# Optimizing kD, c and S by scipy.minimize

```python
drwdwn = B07dd

def drawdown(kD, c, S, drawdwn, tauPS, tauPE):
    Q0 = 4300 * 24 # m3/h
    R = 4700 # m
    tau = drwdwn.T[0]
    rho = R / np.sqrt(kD * c)
    
    #Compute the drawdown
    sH = np.zeros_like(tau)
    for Qp, tauP in zip([Q0, -Q0], [tauPS, tauPE]):
        u = R ** 2 * S / (4 * kD * (tau - tauP)[tau>tauP])
        sH[tau > tauP] += Qp / (4 * np.pi * kD) * Wh(u, rho)
    
    # return residual
    return sH - drwdwn[:, 1]

pars = np.array([kD, c, S])
args = [drwdwn, to_tau(tPS), to_tau(tPE)]

def hant(pars, args):
    return np.sum(drawdown(*pars, *args) ** 2)

res = minimize(hant, pars, args, method='nelder-mead',
              options={'xatol': 1e-8, 'disp': True})
kD, c, S = res['x']
print('kD={:.0f} m2/d, c={:.0f} d, S={:.4g} []'.format(kD, c, S))
```

# Eindverlaging op projectterin door onttrekking Cornelis Biemond. Using $K_0\left(\frac{r}{\lambda}\right)$.

```python
# kD, c, S from uptimization above

r = np.logspace(2, 4, 41)
R = 4700 # m
lam = np.sqrt(kD * c)
Q0 = 4300. * 24 # m3/d

# Steady state Hantush drawdown as function of r and at R
sr = Q0 / (2 * np.pi * kD) * K0(r / lam)
sR = Q0 / (2 * np.pi * kD) * K0(R / lam)

ttl0 = "Drawdown by pumping {:.0f} m3/h at PS CB in Nieuwegein. ".format(Q0 / 24)
ttl1 = "kD = {:.0f} m2/d, c={:.0f} d, S={:4g} [0]".format(kD, c, S)
ax = newfig(ttl0 + ttl1, 'r[ m]', 'Drawdown [m]', xscale='linear')

sH  = 0. # Hantush
sTh = 0. # Theis
for Qp, tauP in zip([Q0, -Q0], [to_tau(tPS), to_tau(tPE)]):
    uMIN  = R ** 2 * S / (4 * kD * (tauMIN - tauP))
    u     = r ** 2 * S / (4 * kD * (tauMIN - tauP))
    sH   += Qp / (4 * np.pi * kD) * Wh(uMIN, R/lam)
    sTh  += Qp / (4 * np.pi * kD) * exp1(u)

ax.plot(r, sr,  label='Hantush steady state')
ax.plot(R, sR, 'o',  mfc='none', label="Steady Hantush at R")
ax.plot(R, sH, '*', mfc='none', label="Transient Hantush at R for t = {:.4g} d".format(tm))
ax.plot(r, sTh, label='Theis for t = {:.4g} d'.format(tm))
ax.plot(R, sMIN, 'v',mfc='none', label="Maximum dd measured")
ax.invert_yaxis()
ax.legend()
```
